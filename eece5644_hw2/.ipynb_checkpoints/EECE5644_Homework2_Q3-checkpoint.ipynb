# Question 3 (35%)
For the datasets you generated in Question 2, implement and apply the Fisher Linear Discriminant Analysis classifier with the decision threshold for the linear discriminant score set to minimize
the smallest probability error you can achieve on the specific data sets generated for each case. Visualize the one-dimensional Fisher LDA discriminant scores and decision labels for each sample
in separate plots for each case.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.linalg import sqrtm
```

The following sample generator code is an exact copy of the one from the 2nd Question

```python
def sample_data_generator(mean1, cov1, mean2, cov2, p1, p2):
    uniform_set = np.random.uniform(0, 1, 400)
    prior_label = [0 if a < p1 else 1 for a in uniform_set]
    z_1 = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], prior_label.count(0))
    z_2 = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], prior_label.count(1))
    data1 = [np.matmul(sqrtm(cov1), z_1[i]) + mean1 for i in range(len(z_1))]
    data2 = [np.matmul(sqrtm(cov2), z_2[i]) + mean2 for i in range(len(z_2))]
    return np.array(data1), np.array(data2)
```

#### Within Class Scatter Matrix
Below I am calculating the Within-Class-Scatter Matrix(Sw).
I first find the mean and covariance from our randomly generated dataset again to make it more accurate.
Then I add both the Covariances to get Sw.

```python
def within_class_scatter_v2(data1, data2):
    print(data1, '\n', data2)
    mean1 = [data1[:, 0].mean(), data1[:, 1].mean()]
    mean2 = [data2[:, 0].mean(), data2[:, 1].mean()]
    cov1 = np.cov(data1.T)
    cov2 = np.cov(data2.T)
#     print(cov1, '\n', cov2)
    Sw = cov1 + cov2
    return Sw
```

#### Between Class Scatter Matrix
Below I am calculating the Between-Class-Scatter Matrix(Sb).
I get the mean again and then calculate Sb

```python
def between_class_scatter_v2(data1, data2):
    mean1 = [data1[:, 0].mean(), data1[:, 1].mean()]
    mean2 = [data2[:, 0].mean(), data2[:, 1].mean()]
    Sb = np.matmul(np.subtract(mean1, mean2).reshape(2, 1), np.transpose(np.subtract(mean1, mean2)).reshape(1, 2))
    return Sb
```

### Calculating the Transofrmed Data points
Below is the main function that calls the functions above.
It is responsible for calculating the Sw and Sb and then using them to generate the Eigen Values and Eigen Vectors.
Once I have the Eigen Values and Vectors, I store them as pairs and sort them based on Eigen Values in Descending Order.

#### Important: I have multiplied the Data points with both the Eigen Vectors to generate the dataset in the same 2-Dimension. I have done so purely for Visualization purposes to demonstrate the shift in the axis of the dataset. Later on while plotting the Fisher LDA, i have only used the first index of dataset which has been transformed in the axis having maximum covariance

Once i prepare my W matrix which is the axis in the direction where the means of both the dataset are the farthest from each other, i transform my dataset in the direction of the new axis and store it in respective Y variables

```python
# Calculating the transformed points in the new W direction and returning them
def calculate_ylda(data1, data2, mean1, mean2):
    Sw = within_class_scatter_v2(data1, data2)
    Sb = between_class_scatter_v2(data1, data2)
    eig_vals, eig_vecs = np.linalg.eig(np.matmul(np.linalg.inv(Sw), Sb))
    eig_vec_value_pair = [(eig_vals[i], eig_vecs[:, i]) for i in range(len(eig_vals))]
    eig_vec_value_pair = sorted(eig_vec_value_pair, key=lambda k: k[0], reverse=True)
    W = np.hstack([eig_vec_value_pair[0][1].reshape(2, 1), eig_vec_value_pair[1][1].reshape(2, 1)])
    Y_lda1 = data1.dot(W)
    Y_lda2 = data2.dot(W)
    return Y_lda1, Y_lda2
```

The function below iterates over multiple threshold values ranging from the minimum possible value of the dataset to the maximum possible value and calculate the Probability of error for each of the threshold.
Using this I am able to calculate the Optimum Threshold value which is the case for which the Probability of error is minimum.

```python
def calculate_min_prob_threshold(data1x, data2x, direction='same'):
    optimum_threshold = min(min(data1x), min(data2x))
    min_error_prob = 1.0
    threshold_error_dict = {}
    for threshold in np.arange(min(min(data1x), min(data2x)), max(max(data1x), max(data2x)), 0.1):
        if direction=='same':
            data1x_classify = [0 if a<threshold else 1 for a in data1x]
            data2x_classify = [0 if b>threshold else 1 for b in data2x]
            error_prob = (data1x_classify.count(1) + data2x_classify.count(1))/(len(data1x) + len(data2x))
        if direction=='different':
            data1x_classify = [0 if a>threshold else 1 for a in data1x]
            data2x_classify = [0 if b<threshold else 1 for b in data2x]
            error_prob = (data1x_classify.count(1) + data2x_classify.count(1))/(len(data1x) + len(data2x))
        threshold_error_dict[threshold] = error_prob
        if error_prob < min_error_prob:
            min_error_prob = error_prob
            optimum_threshold = threshold
    return optimum_threshold, min_error_prob, threshold_error_dict
            
```

## 1.
N=400

For Class Label 1: Mean = [0, 0] Covariance = I (Identity Matrix)

For Class Label 2: Mean = [3, 3] Covariance = I (Identity Matrix)

```python
# Generating samples for Label 1
cov1 = np.array([[1.0, 0.0],
                [0.0, 1.0]])
mean1 = np.array([0., 0.])
p1 = 0.5
```

```python
# Generating samples for Label 2
cov2 = np.array([[1.0, 0.0],
                [0.0, 1.0]])
mean2 = np.array([3., 3.])
p2 = 0.5
```

```python
data1, data2 = sample_data_generator(mean1, cov1, mean2, cov2, p1, p2)
```

```python
# Plot of the original dataset
plt.scatter(data1[:, 0], data1[:, 1], c='#1f77b4', marker='^', label='Label 1')
plt.scatter(data2[:, 0], data2[:, 1], c='#ff7f0e', marker='^', label='Label 2')
plt.legend()
plt.xlabel('X1 (First Dimension)')
plt.ylabel('X2 (Second Dimension)')
plt.title('Scatter Plot of the Bivariate Sample Dataset')
```

```python
y_lda1, y_lda2 = calculate_ylda(data1, data2, mean1, mean2)
```

```python
# Plot of the transformed dataset
plt.scatter(y_lda1[:, 0], y_lda1[:, 1], c='#1f77b4', marker='^', label='Label 1')
plt.scatter(y_lda2[:, 0], y_lda2[:, 1], c='#ff7f0e', marker='^', label='Label 2')
plt.legend()
plt.xlabel('Y1 (First Dimension)')
plt.ylabel('Y2 (Second Dimension)')
plt.title('Scatter Plot of the Bivariate Sample Dataset in the Transformed X Axis')
```

```python
if y_lda2[:, 0].mean() > y_lda1[:, 0].mean():
    direction = 'same'
else:
     direction = 'different'   
optimum_threshold, min_error_prob, threshold_error_dict = calculate_min_prob_threshold(y_lda1[:, 0], 
                                                                                       y_lda2[:, 0], 
                                                                                       direction=direction)
```

```python
plt.plot(list(threshold_error_dict.keys()), list(threshold_error_dict.values()), label='Threshold - Error Curve')
plt.legend()
plt.xlabel('Threshold')
plt.ylabel('Probability of Error')
plt.title('Plot of Threshold Values vs Probability of Error present on each of the threshold values')
```

```python
y_lda1_edge = [('none' if a > optimum_threshold else [0, 0, 0]) if direction == 'different' else ('none' if a < optimum_threshold else [0, 0, 0]) for a in y_lda1[:, 0]]
y_lda2_edge = [('none' if b < optimum_threshold else [0, 0, 0]) if direction == 'different' else ('none' if b > optimum_threshold else [0, 0, 0]) for b in y_lda2[:, 0]]
```

```python
plt.scatter(y_lda1[:, 0], np.zeros([len(y_lda1[:, 0])]), edgecolors=y_lda1_edge, label='Label 1')
plt.scatter(y_lda2[:, 0], np.zeros([len(y_lda2[:, 0])]), edgecolors=y_lda2_edge, label='Label 2')
plt.legend()
plt.xlabel('Maximum Covariance Axis')
plt.title('1D Scatter Plot of the Bivariate Sample Dataset')
```

#### NOTE: Labels with Black Outlines are the Incorrect Classifications in the plot

```python
# Probability of Error
num_misclassified_samples = len([a for a in y_lda1_edge if a != 'none']) + len([a for a in y_lda2_edge if a != 'none'])

# The probability of error here is the number of labels that are classified incorrectly compared to the total 
# number of classified samples which is 400
p_error = num_misclassified_samples / (400)
print("The Probability of error is: {}".format(p_error))
```
