```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import copy
```

Importing the first image

```python
color_list = [[0, 0, 0], [255, 255, 255], [255, 0, 0], [0, 0, 255], [0, 255, 0]]
```

```python
image2_feature
```

KMeans segmentation on the two images

```python
# Normalizing the first feature vector
image1_feature_normalized = np.array(image1_feature)
image1_feature = np.array(image1_feature).astype(float)
for i in range(image1_feature.shape[1]):
    max_i = np.max(image1_feature[:, i])
    for j in range(image1_feature.shape[0]):        
        image1_feature_normalized[j, i] = image1_feature[j, i].astype(float) / max_i
```

```python
image1_feature_normalized
```

```python
# Normalizing the first feature vector
image2_feature_normalized = np.array(image2_feature)
image2_feature = np.array(image2_feature).astype(float)
for i in range(image1_feature.shape[1]):
    max_i = np.max(image2_feature[:, i])
    for j in range(image2_feature.shape[0]):        
        image2_feature_normalized[j, i] = image2_feature[j, i].astype(float) / max_i
```

```python
image2_feature_normalized
```

```python
def kmeans_segmentation(nclusters, feature_vector, norm_feature_vector, image1):
    segmented_image = image1
    km = KMeans(n_clusters=nclusters)
    km.fit(norm_feature_vector)
    cluster_color_dict = {}
    cluster_centers = km.cluster_centers_
    for i in range(nclusters):
        cluster_color_dict[i] = color_list[i]
    for vec in np.hstack([feature_vector, km.labels_[:, np.newaxis]]):
        segmented_image[int(vec[0]), int(vec[1])] = cluster_color_dict[vec[5]]
    return segmented_image
```

```python
def gmm_segmentation(nclusters, feature_vector, norm_feature_vector, image1):
    segmented_image = image1
    gmm = GaussianMixture(n_components=nclusters)
    gmm_labels = gmm.fit_predict(norm_feature_vector)
    cluster_color_dict = {}
    cluster_centers = gmm.means_
    for i in range(nclusters):
        cluster_color_dict[i] = color_list[i]
    for vec in np.hstack([feature_vector, gmm_labels[:, np.newaxis]]):
        segmented_image[int(vec[0]), int(vec[1])] = cluster_color_dict[vec[5]]
    return segmented_image
```
