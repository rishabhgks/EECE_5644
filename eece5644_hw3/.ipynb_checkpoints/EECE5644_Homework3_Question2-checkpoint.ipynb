### 1. Generate training set with 999 2-dimensional samples from two classes with priors q− = 0.3 and q+ = 0.7; 
the class-conditional data probability distributions are two Gaussians with
different mean vectors and different covariance matrices (choose the matrices to be nondiagonal with distinct eigenvalues, so your Gaussian pdfs are tilted with respect to each other
and elongated in different directions by different aspect ratios). Hint: For more interesting
results, make the Gaussians overlap with each other somewhat significantly, so that the
minimum error probability achievable is not too small.

```python
import numpy as np
import copy
from scipy.stats import multivariate_normal
import math
from scipy.linalg import sqrtm
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# Defining Mean and Covariance for 1st component
def comp_one(size):
    cov1 = np.array([[2.0, 1],
                    [1, 1.0]])
    mean1 = np.array([2.0, 2.0])
    return np.random.multivariate_normal(mean=mean1, cov=cov1, size=size)
p1 = 0.3
```

```python
# Defining Mean and Covariance for 1st component
def comp_two(size):
    cov2 = np.array([[2.0, 2],
                    [2, 3.0]])
    mean2 = np.array([-2.0, 2.0])
    return np.random.multivariate_normal(mean=mean2, cov=cov2, size=size)
p2 = 0.7
```

```python
sample_dist = np.random.uniform(0, 1, 999)
count_one = 0
count_two = 0
for sample in sample_dist:
    if sample <= 0.3:
        count_one += 1
    else:
        count_two += 1
print(count_one, count_two)
```

```python
data_part1 = comp_one(count_one)
data_part2 = comp_two(count_two)
dataset = np.array(data_part1.tolist() + data_part2.tolist())
```

### 2. Using Fisher LDA, identify a linear classifier that minimizes the error count on the training set. 
This classifier will have a discriminant function of the form wTLDAx + bLDA where x is
the data vector, and the classifier decides in favor of Class − if the discriminant is below 0,
and decides in favor of Class +, if the discriminant is at least 0.

```python
def within_class_scatter_v2(data1, data2):
    mean1 = [data1[:, 0].mean(), data1[:, 1].mean()]
    mean2 = [data2[:, 0].mean(), data2[:, 1].mean()]
    cov1 = np.cov(data1.T)
    cov2 = np.cov(data2.T)
#     print(cov1, '\n', cov2)
    Sw = cov1 + cov2
    return Sw
```

```python
def between_class_scatter_v2(data1, data2):
    mean1 = [data1[:, 0].mean(), data1[:, 1].mean()]
    mean2 = [data2[:, 0].mean(), data2[:, 1].mean()]
    Sb = np.matmul(np.subtract(mean1, mean2).reshape(2, 1), np.transpose(np.subtract(mean1, mean2)).reshape(1, 2))
    return Sb
```

```python
# Calculating the transformed points in the new W direction and returning them
def calculate_ylda(data1, data2):
    Sw = within_class_scatter_v2(data1, data2)
    Sb = between_class_scatter_v2(data1, data2)
    eig_vals, eig_vecs = np.linalg.eig(np.matmul(np.linalg.inv(Sw), Sb))
    eig_vec_value_pair = [(eig_vals[i], eig_vecs[:, i]) for i in range(len(eig_vals))]
    eig_vec_value_pair = sorted(eig_vec_value_pair, key=lambda k: k[0], reverse=True)
    W = np.hstack([eig_vec_value_pair[0][1].reshape(2, 1), eig_vec_value_pair[1][1].reshape(2, 1)])
    Y_lda1 = data1.dot(W)
    Y_lda2 = data2.dot(W)
    return Y_lda1, Y_lda2
```

```python
def calculate_min_prob_threshold(data1x, data2x, direction='same'):
    optimum_threshold = min(min(data1x), min(data2x))
    min_error_prob = 1.0
    threshold_error_dict = {}
    for threshold in np.arange(min(min(data1x), min(data2x)), max(max(data1x), max(data2x)), 0.1):
        if direction=='same':
            data1x_classify = [0 if a<threshold else 1 for a in data1x]
            data2x_classify = [0 if b>threshold else 1 for b in data2x]
            error_prob = (data1x_classify.count(1) + data2x_classify.count(1))/(len(data1x) + len(data2x))
        if direction=='different':
            data1x_classify = [0 if a>threshold else 1 for a in data1x]
            data2x_classify = [0 if b<threshold else 1 for b in data2x]
            error_prob = (data1x_classify.count(1) + data2x_classify.count(1))/(len(data1x) + len(data2x))
        threshold_error_dict[threshold] = error_prob
        if error_prob < min_error_prob:
            min_error_prob = error_prob
            optimum_threshold = threshold
    return optimum_threshold, min_error_prob, threshold_error_dict
            
```

```python
y_lda1, y_lda2 = calculate_ylda(data_part1, data_part2)
```

```python
# Plot of the transformed dataset
plt.scatter(y_lda1[:, 0], y_lda1[:, 1], c='#1f77b4', marker='^', label='Label 1', s=6)
plt.scatter(y_lda2[:, 0], y_lda2[:, 1], c='#ff7f0e', marker='^', label='Label 2', s=6)
plt.legend()
plt.xlabel('Y1 (First Dimension)')
plt.ylabel('Y2 (Second Dimension)')
plt.title('Scatter Plot of the Bivariate Sample Dataset in the Transformed X Axis')
```

```python
if y_lda2[:, 0].mean() > y_lda1[:, 0].mean():
    direction = 'same'
else:
     direction = 'different'   
optimum_threshold, min_error_prob, threshold_error_dict = calculate_min_prob_threshold(y_lda1[:, 0], 
                                                                                       y_lda2[:, 0], 
                                                                                       direction=direction)
```

```python
plt.plot(list(threshold_error_dict.keys()), list(threshold_error_dict.values()), label='Threshold - Error Curve')
plt.legend()
plt.xlabel('Threshold')
plt.ylabel('Probability of Error')
plt.title('Plot of Threshold Values vs Probability of Error present on each of the threshold values')
```

```python
y_lda1_edge = [('none' if a > optimum_threshold else [0, 0, 0]) if direction == 'different' else ('none' if a < optimum_threshold else [0, 0, 0]) for a in y_lda1[:, 0]]
y_lda2_edge = [('none' if b < optimum_threshold else [0, 0, 0]) if direction == 'different' else ('none' if b > optimum_threshold else [0, 0, 0]) for b in y_lda2[:, 0]]
```

```python
plt.scatter(y_lda1[:, 0], np.zeros([len(y_lda1[:, 0])]), edgecolors=y_lda1_edge, label='Label 1')
plt.scatter(y_lda2[:, 0], np.zeros([len(y_lda2[:, 0])]), edgecolors=y_lda2_edge, label='Label 2')
plt.legend()
plt.xlabel('Maximum Covariance Axis')
plt.title('1D Scatter Plot of the Bivariate Sample Dataset')
```

```python
# Probability of Error
num_misclassified_samples = len([a for a in y_lda1_edge if a != 'none']) + len([a for a in y_lda2_edge if a != 'none'])

# The probability of error here is the number of labels that are classified incorrectly compared to the total 
# number of classified samples which is 400
p_error = num_misclassified_samples / (400)
print("The Probability of error is: {}".format(p_error))
```
